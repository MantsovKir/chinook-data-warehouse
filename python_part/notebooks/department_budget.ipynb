```python
import sys
from pathlib import Path

import pandas as pd

sys.path.append(str(Path.cwd().parent / "python"))

from db_connection import get_database_engine

RAW_DATA_DIR = Path.cwd().parent / "data" / "raw"
```

```python
departments = pd.read_csv(
    RAW_DATA_DIR / 'raw-department.txt',
    sep='-'
)

departments
```

```python
budget_source_1 = pd.read_json(
    RAW_DATA_DIR / 'raw-department-budget.txt', lines=True
)

budget_source_1
```

```python
budget_source_2 = pd.read_json(
    RAW_DATA_DIR / 'raw-department-budget2.txt'
)

budget_source_2
```

```python
budget_source_all = pd.concat([budget_source_1,budget_source_2], ignore_index=True)
budget_source_all
```

```python
merged_departments_budget = pd.merge\
    (budget_source_all, departments, on='department_id', how='left')
merged_departments_budget
```

```python
department_budget_summary = merged_departments_budget.groupby\
    (['department_id', 'department_name'])['budget'].sum().reset_index()
department_budget_summary
```

```python
engine = get_database_engine()
```

```python
engine
```

```python
department_budget_summary.to_sql(
    'department_budget',
    engine,
    schema='stg',
    if_exists='replace',
    index=False
)
```
